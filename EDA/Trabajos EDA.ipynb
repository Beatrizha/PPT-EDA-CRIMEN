# EDA Delitos vs Extranjeros

## Tema
Siempre ha existido por una parte de la población española el debate racista y xenófobo que la mayoría de los actos delictivos son ejecutados por extranjeros, existen varias hipótesis que fomenta este debate y que intentaremos resolver en este EDA.
Si bien es cierto que no tenemos el dato exacto de los actos cometidos por extranjeros, si podemos estudiar los datos según los extanjeros por población, es decir, actos delitivos x porcentaje de habitantes en la comunidad autónoma. 

## Hipótesis


1. **Hipótesis de la estructura de oportunidades**: Los inmigrantes pueden tener más probabilidades de involucrarse en actividades delictivas debido a la falta de oportunidades legales para el avance socioeconómico.
2. **Hipótesis de los enfoques culturales**: Las diferencias culturales pueden llevar a malentendidos o conflictos que resultan en comportamientos delictivos.
3. **Hipótesis de la desorganización social**: Los inmigrantes pueden ser más propensos a la delincuencia si viven en áreas con altos niveles de desorganización social.
4. **Hipótesis de la privación relativa**: La percepción de desigualdad y privación entre los inmigrantes puede llevar a un aumento de la delincuencia.
5. **Hipótesis del control social**: La falta de redes de apoyo social y comunitario puede llevar a los inmigrantes a involucrarse en actividades delictivas

Es importante recordar que la delincuencia es un fenómeno complejo influenciado por una multitud de factores, y la nacionalidad o el estatus de inmigrante de una persona son solo una pequeña parte de la ecuación. La investigación empírica es necesaria para probar estas hipótesis y entender mejor la relación entre la inmigración y la delincuencia.
